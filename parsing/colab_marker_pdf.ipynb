```python
import os
import shutil

# Define source and destination paths
pdf_drive_path = "/content/drive/MyDrive/guides/pdf/"
md_drive_path = "/content/drive/MyDrive/guides/markdown/"
local_temp_pdf_path = "/content/temp_pdfs/"
local_temp_md_path = "/content/temp_markdown/"

# # remove temp folders content
# shutil.rmtree(local_temp_pdf_path)
# shutil.rmtree(local_temp_md_path)

# Create temporary directories
os.makedirs(local_temp_pdf_path, exist_ok=True)
os.makedirs(local_temp_md_path, exist_ok=True)


# Copy PDF files from Drive to local temporary directory
for filename in os.listdir(pdf_drive_path):
    if filename.endswith(".pdf"):
        shutil.copy(os.path.join(pdf_drive_path, filename), local_temp_pdf_path)
print(f"Copied PDF files from {pdf_drive_path} to {local_temp_pdf_path}")

# Run the marker command on the local temporary directory
get_ipython().system(f"marker {local_temp_pdf_path} --workers 1 --output_dir {local_temp_md_path}")

# Copy generated markdown files from local temporary directory to Drive
shutil.copytree(local_temp_md_path, md_drive_path, dirs_exist_ok=True)
print(f"Copied markdown files from {local_temp_md_path} to {md_drive_path}")

```
